## Python exercise: keyspace and crack time


```python
for name, N in {"lowercase":26, "+digits":36, "+symbols":95}.items():
    print(f"{name:10} length 8 -> {N**8:,} combinations")
```

```python
def crack_years(N, L, gps=1_000_000_000):
    return (N ** L) / gps / (60*60*24*365)
for L in [6, 8, 12, 16]:
    print(L, "chars ->", round(crack_years(95, L), 4), "years")
```

## Python exercise: hashing and why salt matters


```python
import hashlib
def h(x): return hashlib.sha256(x.encode()).hexdigest()[:24]
print("same password, no salt -> same hash:")
print(h("squad123"), h("squad123"))
print("with salts -> different hashes:")
print(h("saltA:squad123"), h("saltB:squad123"))
```
