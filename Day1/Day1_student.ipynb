## Fundamentals - Data Types and Operations

### Part 1: Numeric Operations

```python
# Integer arithmetic
# Declare two integer variables and perform all five operations.
a = ___
b = ___

print("Addition:",       a + b)
print("Subtraction:",    a - b)
print("Multiplication:", a * b)
print("Division:",       a / b)   # always returns a float in Python 3
print("Modulus:",        a % b)
```

```python
# Float arithmetic
# Repeat the same operations with float values and compare the results.
x = ___  # e.g. 7.5
y = ___  # e.g. 2.0

print("Float addition:",       x + y)
print("Float subtraction:",    x - y)
print("Float multiplication:", x * y)
print("Float division:",       x / y)
print("Float modulus:",        x % y)

# Note: in Python 3, dividing two ints still gives a float (e.g. 7/2 = 3.5).
# Use // for integer (floor) division.
```

```python
# Data type limits and overflow behaviour
# Python integers have no fixed size — they grow as needed.
# Floats follow IEEE 754 and do have limits.

big_int = 2 ** 100   # Python handles this without overflow
print("Big int:", big_int)

import sys
print("Max float:", sys.float_info.max)

# What happens when a float exceeds its range?
overflow_test = 1.8e308 * 10
print("Float overflow:", overflow_test)  # expect 'inf'
```

### Part 2: Character and String Operations

```python
# Character variables
# Python has no char type; a single character is a length-1 string.

letter = ___  # e.g. 'A'

# Print the ASCII (Unicode) value of the character.
print("Character:", letter)
print("ASCII value:", ord(letter))   # ord() is Python's equivalent of casting char to int
```

```python
# String variables and built-in methods
sentence = ___  # e.g. "Cybersecurity is the practice of protecting systems."

# Length — equivalent to Java's .length()
print("Length:", len(sentence))

# Character at a specific index — equivalent to Java's .charAt(i)
index = ___  # choose any valid index
print(f"Character at index {index}:", sentence[index])

# Substring — equivalent to Java's .substring(start, end)
print("Substring [0:13]:", sentence[0:13])

# Upper and lower case — same method names as Java but no parentheses needed differently
print("Uppercase:", sentence.upper())
print("Lowercase:", sentence.lower())
```
