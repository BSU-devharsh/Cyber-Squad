## Exploring Variables — Solutions

## Teaching note

types at runtime — no explicit type declarations needed. Compare with Java where
students must write `int age = 20;`. Ask: what does `type(age)` return?

## Fundamentals — Solutions

### Part 1: Numeric Operations

## Teaching note
Exercise 7: highlight that Python 3 division always yields float (7/2 = 3.5).
This differs from Java where int/int truncates. Introduce // for floor division.

```python
# Exercise 7 SOLUTION: Integer arithmetic
a = 15
b = 4
print("Addition:",       a + b)   # 19
print("Subtraction:",    a - b)   # 11
print("Multiplication:", a * b)   # 60
print("Division:",       a / b)   # 3.75  (float in Python 3)
print("Modulus:",        a % b)   # 3
```

```python
# Exercise 8 SOLUTION: Float arithmetic
x = 7.5
y = 2.0
print("Float addition:",       x + y)  # 9.5
print("Float subtraction:",    x - y)  # 5.5
print("Float multiplication:", x * y)  # 15.0
print("Float division:",       x / y)  # 3.75
print("Float modulus:",        x % y)  # 1.5
```

```python
# Exercise 9 SOLUTION: Data type limits
big_int = 2 ** 100
print("Big int:", big_int)  # 1267650600228229401496703205376 — no overflow

import sys
print("Max float:", sys.float_info.max)  # ~1.8e+308

overflow_test = 1.8e308 * 10
print("Float overflow:", overflow_test)  # inf
```

### Part 2: Character and String Operations

## Teaching note
Exercise 10: `ord()` returns the Unicode code point. `chr()` is the inverse.
Ask: what does `chr(65)` return? Answer: 'A'. Draw the parallel to Java's char cast.

```python
# Exercise 10 SOLUTION: Character variables
letter = 'A'
print("Character:", letter)
print("ASCII value:", ord(letter))  # 65
```

```python
# Exercise 11 SOLUTION: String variables and built-in methods
sentence = "Cybersecurity is the practice of protecting systems."

print("Length:", len(sentence))           # 51
index = 0
print(f"Character at index {index}:", sentence[index])  # 'C'
print("Substring [0:13]:", sentence[0:13]) # 'Cybersecurity'
print("Uppercase:", sentence.upper())
print("Lowercase:", sentence.lower())
```

## Linux and Networking — Solutions

## Teaching note
Exercise 12 introduces iteration over a list. Students who have not coded before
should be able to run this cell as-is and observe output before modifications.
Ask: what would happen if we add a fourth IP to the list?

```python
# Exercise 12 SOLUTION: Print each host
hosts = ["10.0.0.5", "10.0.0.6", "10.0.0.7"]
for h in hosts:
    print("in-range host:", h)
```

## Teaching note
Exercise 13: the 3 reserved addresses are network address, broadcast address,
and default gateway. Standard IPv4 subnet knowledge.
Ask: what changes for a /25 (128 addresses)?

```python
# Exercise 13 SOLUTION: Subnet math
total = 256
reserved = 3  # network + broadcast + gateway
print("usable hosts:", total - reserved)  # 253
```

## Teaching note
Challenge: enumerate(start=1). Common mistake — students forget the index
variable name or start from 0. Walk through: enumerate returns (index, value) pairs.

```python
# Challenge SOLUTION: enumerate() with start=1
for i, h in enumerate(hosts, 1):
    print(f"host {i}: {h}")
```

## Pair work SOLUTION
Store 5 IPs, print numbered, calculate /26 usable hosts.

```python
# PAIR WORK SOLUTION
hosts5 = ["192.168.1.1","192.168.1.2","192.168.1.3","192.168.1.4","192.168.1.5"]
for i, h in enumerate(hosts5, 1):
    print(f"host {i}: {h}")
# /26: 64 total, 2 reserved (network + broadcast)
print("usable /26 hosts:", 64 - 2)
```
