## Teaching note
Exercise 1 introduces iteration over a list. Students who have not coded before
should be able to run this cell as-is and observe output before any modifications.
Ask: what would happen if we add a fourth IP to the list?

```python
# Exercise 1 SOLUTION: Print each host
hosts = ["10.0.0.5", "10.0.0.6", "10.0.0.7"]
for h in hosts:
    print("in-range host:", h)
```

## Teaching note
Exercise 2 introduces arithmetic. The 3 reserved addresses are: network address,
broadcast address, and default gateway. This is standard IPv4 subnet knowledge.
Ask: what changes for a /25 (128 addresses)?

```python
# Exercise 2 SOLUTION: Subnet math
total = 256
reserved = 3  # network + broadcast + gateway
print("usable hosts:", total - reserved)  # should print 253
```

## Teaching note
The challenge introduces enumerate(start=1). Common mistake: students write
enumerate(hosts, 1) but forget the variable name for the index, or start from 0.
Walk through: enumerate returns (index, value) pairs.

```python
# Challenge SOLUTION: enumerate() with start=1
for i, h in enumerate(hosts, 1):
    print(f"host {i}: {h}")
```

## Pair work SOLUTION
# Store 5 IPs, print numbered, calculate /26 usable hosts

```python
# PAIR WORK SOLUTION
hosts = ["192.168.1.1", "192.168.1.2", "192.168.1.3", "192.168.1.4", "192.168.1.5"]
for i, h in enumerate(hosts, 1):
    print(f"host {i}: {h}")
# /26 subnet: 64 total, 2 reserved (network + broadcast; gateway is inside the range)
print("usable /26 hosts:", 64 - 2)
```
