## Python exercise: variables, loops, and subnet math


```python
hosts = ["10.0.0.5", "10.0.0.6", "10.0.0.7"]
for h in hosts:
    print("in-range host:", h)
```

```python
# A small subnet has 256 total addresses; 3 are reserved.
total, reserved = 256, 3
print("usable hosts:", total - reserved)
```

### Challenge
Write a loop that prints each host with a label like `host 1: 10.0.0.5`. Use enumerate().


```python
for i, h in enumerate(hosts, 1):
    print(f"host {i}: {h}")
```
