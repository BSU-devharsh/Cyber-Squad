## Teaching note
Exercise 1: the flag format flag{...} is standard CTF convention. Students
decode ZmxhZ3tjeWJlcl9zcXVhZF9jaGFtcHN9 to get flag{cyber_squad_champs}.
This is a real CTF skill - base64 decoding is one of the first things you
try when you see a blob of text in a CTF challenge.

```python
# Exercise 1 SOLUTION: Decode Base64 flag
import base64
encoded_flag = "ZmxhZ3tjeWJlcl9zcXVhZF9jaGFtcHN9"
flag = base64.b64decode(encoded_flag).decode()
print(flag)  # flag{cyber_squad_champs}
```

## Teaching note
Exercise 2: multi-step encoding is common in CTFs. Students must recognize
Base64 first, then XOR. The key (13) is the ROT13 key - a fun nod to
another classic encoding. Walk through: what clues tell you to try XOR?
(The bytes look like printable ASCII shifted by a constant offset.)

```python
# Exercise 2 SOLUTION: Multi-step decode
import base64
blob = base64.b64decode("fWRuYk5ZS3Z1PX9SYn5kY3lSfX9icA==")
result = bytes(b ^ 13 for b in blob)
print(result.decode())  # flag{xor_and_b64}
```

## Teaching note
Exercise 3: log triage is a core blue team skill. The pattern - multiple
failed logins followed by a success - is textbook brute force. In a real
SOC this would trigger an alert, the account would be locked, and the source
IP would be investigated. Ask: at what threshold would you lock an account?

```python
# Exercise 3 SOLUTION: Log triage
logs = [
    ("09:02", "alice", "success"),
    ("09:06", "admin", "fail"),
    ("09:06", "admin", "fail"),
    ("09:07", "admin", "fail"),
    ("09:07", "admin", "success"),
]
fails = [t for t in logs if t[2] == "fail"]  # index 2 is status
print("failed attempts:", len(fails))
print("pattern: many fails then success on", fails[0][1], "= brute force")
```

## Teaching note
Bonus: Counter is a dict subclass that counts hashable objects. This pattern
appears in real SIEM tools. The threshold can be tuned - 3 fails in 60
seconds is a common default. Discuss: what are false positives? (A user who
genuinely forgot their password multiple times.)

```python
# Bonus SOLUTION: Count fails per account
from collections import Counter
fail_counts = Counter(t[1] for t in logs if t[2] == "fail")
THRESHOLD = 2
for account, count in fail_counts.items():
    if count > THRESHOLD:
        print(f"ALERT: {account} had {count} failed logins")
```
