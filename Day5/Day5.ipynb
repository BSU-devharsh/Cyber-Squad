## Python exercise: decode a flag and a multi-step flag


```python
import base64
print(base64.b64decode("ZmxhZ3tjeWJlcl9zcXVhZF9jaGFtcHN9").decode())
blob = base64.b64decode("fWRuYk5ZS3Z1PX9SYn5kY3lSfX9icA==")
print(bytes(b ^ 13 for b in blob).decode())
```

## Python exercise: blue team log triage


```python
logs = [
    ("09:02", "alice", "success"),
    ("09:06", "admin", "fail"),
    ("09:06", "admin", "fail"),
    ("09:07", "admin", "fail"),
    ("09:07", "admin", "success"),
]
fails = [t for t in logs if t[2] == "fail"]
print("failed attempts:", len(fails))
print("pattern: many fails then success on", fails[0][1], "= likely brute force")
```

### Protect yourself: strip metadata before posting
Before sharing photos, remove location metadata. On iPhone: Settings > Privacy > Location > Camera > Never. On Android: Camera settings > Save location > Off. Forensic investigators do the reverse, using metadata to place a device at a time and place.


### CTF tip: recognise the encoding first
In a Capture The Flag challenge, always check for Base64 first: look for strings ending in = or ==. Also watch for hex strings (only 0-9 and A-F, or a 0x prefix), ROT13 (all letters and almost readable), and URL encoding (%41 means the letter A). Identifying the encoding is half the solve.


```python
import base64
ct = 'Q1RGe2Jhc2U2NF9pc19ub3RfZW5jcnlwdGlvbn0='
print(base64.b64decode(ct).decode())   # base64 is encoding, not encryption
```
