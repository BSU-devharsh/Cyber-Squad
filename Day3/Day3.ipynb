## Python exercise: Caesar cipher and brute force


```python
def caesar(t, s):
    o = ""
    for c in t:
        if c.isalpha():
            b = 65 if c.isupper() else 97
            o += chr((ord(c)-b+s) % 26 + b)
        else: o += c
    return o
ct = caesar("Defend the range", 5)
print("cipher:", ct)
for s in range(1, 26):
    print(s, caesar(ct, -s))
```

## Python exercise: Base64, XOR, and hashing


```python
import base64, hashlib
m = "meet at noon"
enc = base64.b64encode(m.encode()).decode()
print("base64:", enc, "->", base64.b64decode(enc).decode())
def xor(d,k): return bytes(b ^ k for b in d)
print("xor back:", xor(xor(b"flag{x}",42),42).decode())
print("sha256:", hashlib.sha256(b"password").hexdigest()[:20])
```
