## Teaching note
Exercise 1: KHOOR is HELLO shifted by 3 (Caesar cipher). Students should see
shift 3 produce HELLO. Point out: the computer tries all 25 options in
microseconds. An attacker does the same. Caesar provides zero real security.

```python
# Exercise 1 SOLUTION: Caesar brute force
ciphertext = "KHOOR"
for shift in range(1, 26):  # 1 through 25
    decoded = ""
    for c in ciphertext:
        if c.isalpha():
            decoded += chr((ord(c) - shift - 65) % 26 + 65)
        else:
            decoded += c
    print(f"shift {shift}: {decoded}")  # shift 3 gives HELLO
```

## Teaching note
Exercise 2: Base64 is NOT encryption. Anyone can decode it with zero key.
It is a way to represent binary data as printable text. Used in email
attachments, JWTs, and data URLs. Ask: is encoded the same bytes as msg? No -
it is longer and uses only A-Z, a-z, 0-9, +, /, = characters.

```python
# Exercise 2 SOLUTION: Base64
import base64
msg = b"Hello, Cyber Squad!"
encoded = base64.b64encode(msg)
print("encoded:", encoded)
decoded = base64.b64decode(encoded)
print("decoded:", decoded)
```

## Teaching note
Exercise 3: XOR is its own inverse - XOR-ing twice returns the original.
Single-byte XOR is trivially breakable (only 256 keys) but the concept scales
to AES in stream mode. Students often find the self-inverse property surprising.
Run it and verify back == plaintext.

```python
# Exercise 3 SOLUTION: XOR cipher
key = 42
plaintext = b"Secret"
xored = bytes(b ^ key for b in plaintext)
print("XORed:", xored)
back = bytes(b ^ key for b in xored)
print("recovered:", back)  # should equal plaintext
```

## Teaching note
Exercise 4: count the hex digits that differ between h1 and h2. Usually more
than 32 of the 64 hex chars differ from a single letter change. This is the
avalanche effect and is what makes SHA-256 tamper-evident in digital signatures
and file integrity verification.

```python
# Exercise 4 SOLUTION: Avalanche effect
import hashlib
h1 = hashlib.sha256(b"password").hexdigest()
h2 = hashlib.sha256(b"Password").hexdigest()
print("password hash:", h1)
print("Password hash:", h2)
diff = sum(a != b for a, b in zip(h1, h2))
print(f"{diff} of 64 hex chars differ")
```

## Pair work SOLUTION - XOR cipher function

```python
# PAIR WORK SOLUTION
import base64

def xor_cipher(message, key):
    return bytes(b ^ key for b in message)

# Verify self-inverse property
original = b"Hello"
encrypted = xor_cipher(original, 42)
decrypted = xor_cipher(encrypted, 42)
print("original: ", original)
print("encrypted:", encrypted)
print("decrypted:", decrypted)
assert decrypted == original, "XOR cipher failed self-inverse test"
print("Self-inverse test passed!")

# CyberChef Base64 result (for reference)
msg = "Hello, Cyber Squad!"
b64 = base64.b64encode(msg.encode()).decode()
print("Base64 of 'Hello, Cyber Squad!':", b64)
# SGVsbG8sIEN5YmVyIFNxdWFkIQ==
```
